```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
health = pd.read_csv('../data/health.csv')
```

```python
health.head(3)
```

```python
plt.scatter(health.weight, health.time)
plt.title('Scatter plot of (weight, time)')
plt.xlabel('weight')
plt.ylabel('time')
```

```python
r = np.corrcoef(health.weight, health.time)
r
# 피어슨의 상관계수를 구한다. 1이 아닌 값 중 하나만 취하면됨
```
