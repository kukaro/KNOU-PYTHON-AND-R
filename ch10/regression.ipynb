```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
health = pd.read_csv('../data/health.csv')
```

```python
X_cols = ['weight', 'pulse', 'muscle', 'quarter']
```

```python
X = health[X_cols]
X
```

```python
X = sm.add_constant(X)
X
```

```python
Y = health['time']
Y.head(5)
```

```python
model = sm.OLS(Y, X)
# 최소자승법(OLS)
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
Y_predict = result.predict(X)
```
