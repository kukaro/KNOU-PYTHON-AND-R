```r
hwy_plot <- ggplot(data = mpg) +
geom_point(
    mapping = aes(
        x=displ,
        y=hwy,
        color=class
    )
)
hwy_plot
```

```r
hwy_plot +
geom_smooth(
    mapping = aes(x=displ, y=hwy),
    method='loess'
)
```

```r
hwy_plot +
geom_smooth(
    mapping = aes(x=displ, y=hwy, linetype=drv),
    method='loess'
)
```

```r
hwy_plot +
facet_grid(drv ~ cyl)
```
